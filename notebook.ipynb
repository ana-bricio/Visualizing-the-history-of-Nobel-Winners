**Task -** What is the most commonly awarded gender and birth country?

```python
#Read the .csv file
data = pd.read_csv("data/nobel.csv")
#Based on the number of missing values, it was determined to fill them with 'Unknown'
data = data.fillna("Unknown")

#New method learned - .idxmax "Returns index of first occurrence of maximum over requested axis" 
top_country = data['birth_country'].value_counts().idxmax()
#Filter dataframe based on top country
data_filtered = data[data['birth_country'] == top_country]

top_gender = data_filtered['sex'].value_counts().idxmax()
```

**Task -** Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?

```python
#First we need to divide the years by decades
#New method learned - np.floor() "Rounds the float to the nearest integer"

data["Decade"] = np.floor(data["year"] / 10)
data["Decade"] = data["Decade"] * 10
data["Decade"] = data["Decade"].astype('int')

#Grouping the dataframe by decades, displaying only the countries and finding the proportion
data_propor = data.groupby('Decade', as_index=False)['birth_country'].value_counts(normalize=True)

#Slicing the dataframe to only visualize the USA proportion
data_propor_usa = data_propor[data_propor['birth_country'] == top_country]

#Obtain the decade with the highest proportion
#Slice the database 
max_usa = data_propor_usa[data_propor_usa['proportion'] == data_propor_usa['proportion'].max()]
max_decade_usa = max_usa['Decade'].values[0]
max_decade_usa

#graph = sns.relplot(kind='line', data=data_propor_usa, x='Decade', y='proportion', markers=True)
#graph.set_titles("USA")
```

**Task -** Which decade and Nobel Prize category combination had the highest proportion of female **laureates?**

```python
# Calculate the proportion of females per decade
data['female_winner'] = data['sex'] == 'Female'
data_female_grouped = data.groupby(['Decade', 'category'], as_index=False)['female_winner'].mean()

# The dataframe must be grouped by decade and category, counting its values 
# data_female_grouped = data_female.groupby(['Decade','category'],as_index=False).mean()

# The decade and category are stored as separate variables
max_decade_female = data_female_grouped[data_female_grouped['female_winner'] == data_female_grouped['female_winner'].max()]['Decade'].values[0]
max_category_female = data_female_grouped[data_female_grouped['female_winner'] == data_female_grouped['female_winner'].max()]['category'].values[0]

# The dictionary is created
max_female_dict = {max_decade_female:max_category_female}
```
